```python
import numpy as np


def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
AND(0,0)  #0
AND(1,0)  #0
AND(0,1)  #0
AND(1,1)  #1

#가중치와 편향도입
import numpy as np
x = np.array([0, 1]) #입력
w = np.array([0.5, 0.5]) #가중치
b = -0.7 #편향
w*x  #array([ 0., 0.51)
np.sum(w*x) #0.5
np.sum(w*x) + b # -0.19999999999999996 대략 -0.2 (부동소수점 수에 의한 연산 오차)

#가중치와 편향구하기
def AND (x1, x2):
  x = np.array([x1, x2])
  w = np.array([0.5, 0.5])
  b = -0.7
  tmp = np.sum(w*x) + b
  if tmp <= 0:
    return 0
  else:
    return 1
```

# ch02/nand_gate.py

```python
import numpy as np

# NAND 게이트 구현하기
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5]) #AND와 가중치만다름 (w,b)
    b = 0.7
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

# ch02/or_gate.py

```python
# OR 게이트 구현하기
import numpy as np

def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5]) #AND와 가중치만다름 (w,b)
    b = -0.2
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

```

# ch02/xor_gate.py

```python
# XOR 게이트 구현하기
def XOR(x1, x2):
    s1 = NAND(x1, x2)
    s2 = OR(x1, x2)
    y = AND(s1, s2)
    return y

XOR(0, 0) #0
XOR(1, 0) #1
XOR(0, 1) #1
XOR(1, 1) #0
```
